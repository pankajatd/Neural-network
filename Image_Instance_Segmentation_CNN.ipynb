```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image  =plt.imread('/content/lenna_JPG.jpg')
plt.imshow(image)
```

```python
image.shape
```

```python
#Convert RGB to Gray scale Image
from skimage.color import rgb2gray
gray_image =rgb2gray(image)
plt.imshow(gray_image,cmap='gray')
```

```python
gray_image.shape
```

```python
gray_r =gray_image.reshape(gray_image.shape[0]*gray_image.shape[1])#This done to flatten the image and get as a single array
gray_r
```

```python
gray_r.shape
```

#Semantic Segmentation

```python
for i in range(gray_r.shape[0]):
  #gray_r.mean()is the Threshld value
  if gray_r[i]>gray_r.mean():
    #object/foreground
    gray_r[i] =1
  else:
    #background
    gray_r[i] =0
grey_semantic_segentation = gray_r.reshape(gray_image.shape[0], gray_image.shape[1])

```

```python
plt.imshow(grey_semantic_segentation,cmap='gray')
plt.savefig("grey_semantic_segentation.jpeg")
```

```python
gray_image =rgb2gray(image)
gray_r =gray_image.reshape(gray_image.shape[0]*gray_image.shape[1])#This done to flatten the image and get as a single array
gray_r
```

```python
#Instance Segmentation

for i in range(gray_r.shape[0]):
  #gray_r.mean()is the Threshld value
  if gray_r[i]>gray_r.mean():
    gray_r[i] =3
  elif gray_r[i]>0.5:
    gray_r[i] =2
  elif gray_r[i]>0.25:
    gray_r[i] =1
  else:
    gray_r[i] =0
grey_instance_segentation = gray_r.reshape(gray_image.shape[0], gray_image.shape[1])
```

Avantages of using above Approaches

*   Calculations are ver simple
*   Faster Operation speed
*   When the object and background have high contrast in the image (pixel values vary a lot),this method perofrms really well.

Limitations:


*   If there is no significant gray scale difference or there is an overlap of the grayscale pixel values,it becomes very difficult in those scenarios to get accurate results.
*   In the above case Edge Detection Segmentation is done using Horizontal and Vertical filters.
